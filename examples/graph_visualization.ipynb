```python
print('ResNet-50')
Graph(torchvision.models.resnet50()).visualize(node_size=100)

print('Visual Transformer')
model = Network(C=128, num_classes=1000, genotype=ViT, n_cells=12, preproc=False, C_mult=1)
Graph(model).visualize(node_size=100)

print('DARTS')
model = Network(C=48, num_classes=1000, genotype=DARTS, n_cells=14, preproc=True, C_mult=2)
Graph(model).visualize(node_size=50)

print('VGG16-BN')
Graph(torchvision.models.vgg16_bn()).visualize(node_size=50)

print('MobileNet')
Graph(torchvision.models.mobilenet_v2()).visualize(node_size=50)

print('DenseNet-201')
Graph(torchvision.models.densenet201()).visualize(node_size=30)

print('legend')
Graph(torch.nn.Conv2d(3, 16, 3)).visualize(node_size=200, figsize=(20,3), with_labels=True, vis_legend=True, label_offset=0.0005)
```
